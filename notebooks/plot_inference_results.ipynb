# Visualize DiffDock Inference Results
Load ranked SDF outputs from DiffDock inference and render poses (optionally with a receptor PDB) using the helper in `utils.visualise.view_inference_results()`.

```python
import sys
from pathlib import Path

# Make project package importable (assumes this notebook is in notebooks/)
project_root = Path('..').resolve()
if str(project_root) not in sys.path:
    sys.path.insert(0, str(project_root))
project_root

from pathlib import Path
from utils.visualise import view_inference_results
```

```python


# Point these to your inference outputs
ligand_sdf = Path('../results/user_predictions_small/1a0q/rank1.sdf').resolve()
receptor_pdb = Path('../data/1a0q/1a0q_protein_processed.pdb').resolve()

ligand_sdf2 = Path('../results/user_predictions_small/1a0q/rank2.sdf').resolve()

print(f"Ligand SDF:   {ligand_sdf}  (exists={ligand_sdf.exists()})")
print(f"Receptor PDB: {receptor_pdb}  (exists={receptor_pdb.exists()})")
```

```python
# Render protein + ligand together
# model_indices: list (e.g., [0]) to pick specific poses, or None for all
viewer = view_inference_results(
    results_dir="../results/test_run/1a0q",
    model_indices=[0],
    width=900,
    height=650,
    show_surface=False,       # set True for translucent protein surface
    zoom_to_ligand=True,      # centre camera on the binding site
)
viewer.show()

viewer2 = view_inference_results(
    results_dir="../results/test_run/1a0q_custom",
    model_indices=[0,2,3,4],
    width=900,
    height=650,
    show_surface=False,       # set True for translucent protein surface
    zoom_to_ligand=True,      # centre camera on the binding site
)
viewer2.show()
```
